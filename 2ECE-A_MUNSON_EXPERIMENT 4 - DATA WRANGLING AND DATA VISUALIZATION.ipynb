# Problem 1

#### For this first problem, based on the dataset boards2.csv, we are required to make two data frames and visuals. 

##### The first prompt is Instru = [“Name”, “GEAS”, “Electronics >70”];  where track is constant as Instrumentation and hometown Luzonn 

```python
#Filter the data; equate data to arrange constant variables
instru_sort = df[(df['Track']=='Instrumentation') & (df['Hometown']=='Luzon') & (df['Electronics']>70)]
#Print only the needed columns
Instru = instru_sort[['Name', 'GEAS', 'Electronics']]
Instru
```

#### The second prompt is Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
#Get the general average among the required components of data
df['Average'] = ((df['Math'])+(df['Electronics'])+(df['GEAS'])+(df['Communication']))/4
#Filter the data; equate data to arrange constant variables
Mindy = df[(df['Hometown']=='Mindanao') & (df['Gender']=='Female') & (df['Average']>=55)]
#Print only the needed columns
Mindy = Mindy[['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

# Problem 2

#### This second problem requires us to create a visualization of the different features which contribute to average grades. 

#### Firstly, we will take into consideration the effect that gender has on the overall average grades attained by students:

```python
#Set the size of the visual
plt.figure(figsize=(6, 3))
#Input the data to be displayed
sns.boxplot(data=df, x='Track', y='Average', hue='Gender')
#Label the y axis
plt.ylabel('Average Grade Attained')
#Label the x axis
plt.xlabel('Track of Choice')
#Give a title for the visual
plt.title('Average grades attained by students of different tracks and gender')
plt.show()
```

#### Now we take into consideration hometown, in which we will make a similar overall visual, but with hometown rather than gender:

```python
#Set the size of the visual
plt.figure(figsize=(6, 3))
#Input the data to be displayed
sns.boxplot(data=df, x='Track', y='Average', hue='Hometown')
#Label the y axis
plt.ylabel('Average Grade Attained')
#Label the x axis
plt.xlabel('Track of Choice')
#Give a title for the visual
plt.title('Average grades attained by students of different tracks and hometowns')
plt.show()
```
